##### **Tokenizer**
å¦‚åŒå…¶ä»– ML æ¨¡å‹ã€NLP ä»»å‹™ï¼Œè™•ç†æ–‡å­—è³‡æ–™æ™‚ï¼Œæ¨¡å‹ç„¡æ³•ç›´æ¥è™•ç†åŸå§‹è³‡æ–™ï¼Œæˆ‘å€‘éœ€è¦é€²è¡Œé è™•ç†å°‡æ–‡æœ¬è½‰æ›ç‚ºæ¨¡å‹å¯ä»¥ç†è§£ï¼ˆåƒï¼‰çš„æ•¸å­—æ ¼å¼ï¼Œæ‰€ä»¥éœ€è¦ä½¿ç”¨ **åˆ†è©å™¨ï¼ˆtokenizerï¼‰**

**Tokenizer è² è²¬**
1. å°‡æ–‡æœ¬é€²è¡Œåˆ†\æ–·è©ï¼Œå¦‚æœè™•ç†éæ–‡å­—è³‡æ–™ï¼Œå¯ä»¥æƒ³åƒæˆ jieba æ–·è©
2. å°‡æ–·è©çµæœè½‰æ›ç‚ºæ•¸å­—åºåˆ—ï¼ˆé€éå­—å…¸ï¼‰
3. æ·»åŠ ç‰¹æ®Šç¬¦è™Ÿï¼Œï¼ˆåœ¨æ­¤å…ˆä¸æ·±å…¥è¨è«–ï¼‰

åœ¨ transformers ä¸­ä½¿ç”¨ `AutoTokenizer` è¼‰å…¥åˆ†è©å™¨ï¼Œæ¯å€‹æ¨¡å‹éƒ½æœƒæœ‰å„è‡ªä¸åŒçš„æ–·è©æ¼”ç®—æ³•ä»¥åŠ mapping åˆ°æ•¸å­—åºåˆ—çš„å­—å…¸ï¼Œ`AutoTokenizer` ä½¿ç”¨ `.from_pretrained()` æ–¹æ³•é€éå‚³å…¥çš„ checkpoint è‡ªå‹•ç²å–å°æ‡‰çš„åˆ†è©å™¨ã€‚

```python
from transformers import AutoTokenizer

checkpoint = "google-bert/bert-base-chinese"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

```python
docs = "ä¸åƒè‘¡è„ä¸åè‘¡è„çš®"
tokenized_result = tokenizer(docs)
print(f"æ–·è©çµæœ(æ•¸å­—åºåˆ—)ï¼š {tokenized_result['input_ids']}")
print(f"å°‡æ•¸å­—åºåˆ—é‚„åŸï¼š{tokenizer.decode(tokenized_result['input_ids'])}")
```

##### **Model**
èˆ‡è¼‰å…¥åˆ†è©å™¨çš„æ–¹æ³•ä¸€æ¨£ï¼Œåˆ©ç”¨ `AutoModel` ä»¥ `.from_prerained()` è¼‰å…¥å¸Œæœ›çš„æ¨¡å‹

```python
from transformers import AutoModel
model = AutoModel.from_pretrained(checkpoint)
```

```python
inputs = tokenizer(docs, 
                   return_tensors='pt')
outputs = model(**inputs)
```
